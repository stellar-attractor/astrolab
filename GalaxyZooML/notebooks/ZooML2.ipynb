```python
import tensorflow as tf
import tensorflow_datasets as tfds
import matplotlib.pyplot as plt
from tqdm import tqdm
import numpy as np

# === –ü–∞—Ä–∞–º–µ—Ç—Ä—ã ===
IMG_SIZE = (128, 128)
BATCH_SIZE = 64
AUTOTUNE = tf.data.AUTOTUNE
EPOCHS = 10

# === –ü—Ä–µ–¥–æ–±—Ä–∞–±–æ—Ç–∫–∞ ===
def preprocess(example):
    image = tf.image.resize(example["image"], IMG_SIZE)
    image = tf.cast(image, tf.float32) / 255.0
    table = example["table1"]
    votes = [
        table["t01_smooth_or_features_a01_smooth_fraction"],
        table["t01_smooth_or_features_a02_features_or_disk_fraction"],
        table["t01_smooth_or_features_a03_star_or_artifact_fraction"],
    ]
    label = tf.argmax(tf.stack(votes), axis=0)
    return image, label

# === –ó–∞–≥—Ä—É–∑–∫–∞ –¥–∞—Ç–∞—Å–µ—Ç–∞ ===
print("üîΩ –ó–∞–≥—Ä—É–∂–∞–µ–º Galaxy Zoo 2...")
ds_full = tfds.load("galaxy_zoo2", split="train", shuffle_files=True)

# –ü–æ–¥—Å—á—ë—Ç —Ä–∞–∑–º–µ—Ä–∞ (–µ–¥–∏–Ω–æ–∂–¥—ã ‚Äî –º–µ–¥–ª–µ–Ω–Ω–æ)
print("üì¶ –ü–æ–¥—Å—á—ë—Ç —á–∏—Å–ª–∞ —ç–ª–µ–º–µ–Ω—Ç–æ–≤...")
n_total = 0
for _ in tqdm(ds_full, desc="–ü–æ–¥—Å—á—ë—Ç –∑–∞–ø–∏—Å–µ–π"):
    n_total += 1

train_size = int(n_total * 0.8)
test_size = n_total - train_size

print(f"üî¢ –í—Å–µ–≥–æ: {n_total} | –û–±—É—á–µ–Ω–∏–µ: {train_size} | –¢–µ—Å—Ç: {test_size}")

# === –†–∞–∑–¥–µ–ª–µ–Ω–∏–µ –Ω–∞ train/test ===
ds_train_raw = ds_full.take(train_size)
ds_test_raw  = ds_full.skip(train_size)

def prepare(ds, desc=""):
    ds = ds.map(preprocess, num_parallel_calls=AUTOTUNE)
    ds = ds.shuffle(1000).batch(BATCH_SIZE).prefetch(AUTOTUNE)
    return ds

ds_train = prepare(ds_train_raw, "Train")
ds_test  = prepare(ds_test_raw, "Test")

# === –ú–æ–¥–µ–ª—å ===
model = tf.keras.Sequential([
    tf.keras.layers.Conv2D(32, 3, activation='relu', input_shape=(128, 128, 3)),
    tf.keras.layers.MaxPooling2D(),
    tf.keras.layers.Conv2D(64, 3, activation='relu'),
    tf.keras.layers.MaxPooling2D(),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(3, activation='softmax')
])

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# === –û–±—É—á–µ–Ω–∏–µ ===
print("üöÄ –û–±—É—á–∞–µ–º –º–æ–¥–µ–ª—å...")
history = model.fit(ds_train, validation_data=ds_test, epochs=EPOCHS)

# === –û—Ü–µ–Ω–∫–∞ ===
print("üìä –û—Ü–µ–Ω–∫–∞ –Ω–∞ —Ç–µ—Å—Ç–æ–≤–æ–º –Ω–∞–±–æ—Ä–µ:")
test_loss, test_acc = model.evaluate(ds_test)
print(f"Test Accuracy: {test_acc:.4f}")

# === –ì—Ä–∞—Ñ–∏–∫–∏ ===
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], label='Train')
plt.plot(history.history['val_loss'], label='Val')
plt.title("Loss")
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['accuracy'], label='Train')
plt.plot(history.history['val_accuracy'], label='Val')
plt.title("Accuracy")
plt.legend()
plt.tight_layout()
plt.show()

# === –ü—Ä–∏–º–µ—Ä—ã –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–π ===
class_names = ["Elliptical", "Spiral", "Artifact"]
for images, labels in ds_test.take(1):
    preds = model.predict(images)
    preds_labels = np.argmax(preds, axis=1)

    plt.figure(figsize=(10, 10))
    for i in range(9):
        plt.subplot(3, 3, i+1)
        plt.imshow(images[i])
        plt.title(f"Pred: {class_names[preds_labels[i]]}\nTrue: {class_names[labels[i]]}")
        plt.axis("off")
    plt.tight_layout()
    plt.show()
```

```python
# üì¶ –ò–º–ø–æ—Ä—Ç
import os
import csv
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import load_model
from PIL import Image
from tqdm import tqdm

# üß† –ó–∞–≥—Ä—É–∑–∫–∞ –º–æ–¥–µ–ª–∏
model = load_model("galaxy_cnn_model.keras")
class_names = ["Elliptical", "Spiral", "Artifact"]

# üìÇ –ü–∞–ø–∫–∞ —Å –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è–º–∏
image_dir = "new_galaxy_images"
output_csv = "predictions.csv"

# ‚öôÔ∏è –ü–∞—Ä–∞–º–µ—Ç—Ä—ã –æ–±—Ä–∞–±–æ—Ç–∫–∏
img_size = (128, 128)  # —Ä–∞–∑–º–µ—Ä, –∫–æ—Ç–æ—Ä—ã–π –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–ª—Å—è –ø—Ä–∏ –æ–±—É—á–µ–Ω–∏–∏

# üñºÔ∏è –ó–∞–≥—Ä—É–∑–∫–∞ –∏ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–µ
results = []

for filename in tqdm(sorted(os.listdir(image_dir)), desc="üîç –ö–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏—è"):
    if filename.lower().endswith((".jpg", ".jpeg", ".png")):
        path = os.path.join(image_dir, filename)
        try:
            img = Image.open(path).convert("RGB").resize(img_size)
            img_array = np.array(img) / 255.0
            img_array = np.expand_dims(img_array, axis=0)
            prediction = model.predict(img_array, verbose=0)
            predicted_label = class_names[np.argmax(prediction)]
            results.append((filename, predicted_label))
        except Exception as e:
            print(f"‚ö†Ô∏è –û—à–∏–±–∫–∞ —Å —Ñ–∞–π–ª–æ–º {filename}: {e}")

# üíæ –°–æ—Ö—Ä–∞–Ω–µ–Ω–∏–µ –≤ CSV
with open(output_csv, "w", newline="") as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(["Filename", "Predicted Class"])
    writer.writerows(results)

print(f"‚úÖ –ö–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏—è –∑–∞–≤–µ—Ä—à–µ–Ω–∞. –†–µ–∑—É–ª—å—Ç–∞—Ç—ã –≤ {output_csv}")
```
